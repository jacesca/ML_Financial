# 1. SEABRON SCATTERPLOT & COUNTPLOT

```
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import pandas as pd # Import Pandas for data manipulation using dataframes
import numpy as np # Import Numpy for data statistical analysis 
import matplotlib.pyplot as plt # Import matplotlib for data visualisation
import seaborn as sns # Statistical data visualization
```

```
# np.C_ class object translates slice objects to concatenation along the second axis.
x1 = np.array([1,2,3])
x1.shape
```

```
x2 = np.array([4,5,6])
x2.shape
```

```
z = np.c_[x1, x2]
print(z.shape)
z
```

```
z = np.vstack((x1, x2)).T
print(z.shape)
z
```

```
# Import Cancer data drom the Sklearn library
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```
cancer
```

```
np.c_[cancer['data'], cancer['target']]
```

```
# Create a dataFrame named df_cancer with input/output data
# # First method
# df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns=np.append(cancer['feature_names'], ['target']))

# # Second method
# df_cancer = pd.DataFrame(np.hstack((cancer['data'], cancer['target'].reshape(len(cancer['target']),1))), columns=np.append(cancer['feature_names'], ['target']))

# # Third method
df_cancer = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df_cancer['target'] = cancer['target']
df_cancer.info()
```

```
# Check out the head of the dataframe
df_cancer.head()
```

```
# Check out the tail of the dataframe
df_cancer.tail()
```

```
%matplotlib inline
# Plot scatter plot between mean area and mean smoothness

sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df_cancer);
plt.title('Breast cancer file');
```

```
%matplotlib inline

# Let's print out countplot to know how many samples belong to class #0 and #1
sns.countplot(x=df_cancer['target'], label = "Count") 
plt.title('Breast cancer file');
```

# 2. SEABORN PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS

```
%matplotlib inline

# Plot the pairplot
sns.pairplot(df_cancer, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
plt.suptitle('Breast cancer file');
plt.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=.9, top=.95, wspace=None, hspace=None);
```

```
%matplotlib inline

# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
plt.figure(figsize = (20, 10)) 
sns.heatmap(df_cancer.corr(), annot = True) 
plt.suptitle('Breast cancer file')
plt.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=.95, wspace=None, hspace=None);
```

```
%matplotlib inline

# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 

sns.histplot(df_cancer['mean radius'], bins = 25, color = 'blue', kde=True, linewidth=0)
plt.suptitle('Breast cancer file')
plt.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=.9, wspace=None, hspace=None);
```

**MINI CHALLENGE #2:**
- **Plot two separate distplot for each target class #0 and target class #1**


```
%matplotlib inline

plt.figure(figsize = (20, 5))

plt.subplot(1, 2, 1)
sns.histplot(df_cancer[df_cancer.target==0]['mean radius'], bins = 25, color = 'blue', kde=True, linewidth=0)
plt.title('Target 0')
plt.grid()

plt.subplot(1, 2, 2)
sns.histplot(df_cancer[df_cancer.target==1]['mean radius'], bins = 25, color = 'blue', kde=True, linewidth=0)
plt.title('Target 1')
plt.grid()

plt.suptitle('Breast cancer file')
plt.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=.87, wspace=None, hspace=None);
```

```
%matplotlib inline

plt.figure(figsize = (10, 5))
sns.histplot(df_cancer[df_cancer.target==0]['mean radius'], bins = 25, color = 'green', kde=True, linewidth=0)
sns.histplot(df_cancer[df_cancer.target==1]['mean radius'], bins = 25, color = 'red', kde=True, linewidth=0)
plt.grid()
plt.suptitle('Breast cancer file')
plt.tight_layout()
plt.subplots_adjust(left=None, bottom=None, right=None, top=.9, wspace=None, hspace=None);
```
